# 2. Variable Validation

## General Variables

#### ODATEDW

```python
working_data['ODATEDW'] =  pd.to_datetime(data['ODATEDW'], format='%Y-%m-%d') # Converting it to datetime
```

```python
#current_date = date(2020,11,30)
current_date = date.today() # sets current date
print(type(current_date), current_date)
```

```python
working_data['0month'] = working_data['ODATEDW'].apply(lambda x : x.month)
working_data['0year'] = working_data['ODATEDW'].apply(lambda x : x.year)
working_data['contr_since'] = working_data['ODATEDW'].dt.date.apply(lambda x : (current_date-x).days)
```

#### OSOURCE

```python
working_data.drop(columns='OSOURCE', inplace=True) # nominal value, can't do much with it. 
```

#### TCODE

```python
working_data.drop(columns='TCODE', inplace=True) # nominal value, can't do much with it. 
```

#### ZIP 

```python
# there are sometimes redundant symbols in the end of ZIP code, we need 1st 5
working_data['ZIP'] = working_data['ZIP'].apply(lambda x: int(x[0:4]))
print('NaNs: ', sum(working_data['ZIP'].isna()))

working_data['ZIP'].hist(bins = 10);
```

#### MAILCODE

```python
# ' ' initially indicated that addres was OK 
display(working_data['MAILCODE'].unique())

working_data['MAILCODE'] = working_data['MAILCODE'].apply(lambda x: 1 if x == ' ' else 0)

print('NaNs: ', sum(working_data['MAILCODE'].isna()))
working_data['MAILCODE'].value_counts()
```

#### PVASTATE 

```python
display(working_data['PVASTATE'].unique())

 # let's turn it into binary
working_data['PVASTATE'] = working_data['PVASTATE'].apply( lambda x: 0 if x == ' ' else 1)

display(working_data['PVASTATE'].value_counts())
print('NaNs:', working_data['PVASTATE'].isna().sum())
```

#### DOB 

```python
working_data['DOB'] =  pd.to_datetime(working_data['DOB'], format='%Y-%m-%d')
```

```python
working_data['born_day'] = working_data['ODATEDW'].apply(lambda x : x.day)
working_data['born_month'] = working_data['ODATEDW'].apply(lambda x : x.month)
working_data['born_year'] = working_data['ODATEDW'].apply(lambda x : x.year)
```

#### NOEXCH, RECINHSE, RECP3, RECPGVG, RECSWEEP - converted to binary

```python
display('NOEXCH', working_data['NOEXCH'].unique())
display('RECINHSE', working_data['RECINHSE'].unique())
display('RECP3', working_data['RECP3'].unique())
display('RECPGVG', working_data['RECPGVG'].unique())
display('RECSWEEP', working_data['RECSWEEP'].unique())
```

```python
working_data['NOEXCH'] = working_data['NOEXCH'].replace({'0':0, '1':1, 'X':1, ' ':0}).astype(int)
print(working_data[['NOEXCH']].value_counts())
print('NaNs:', working_data['NOEXCH'].isna().sum())

working_data['RECINHSE'] = working_data['RECINHSE'].apply(lambda x: 1 if x=='X' else 0)
print(working_data[['RECINHSE']].value_counts())
print('NaNs:', working_data['RECINHSE'].isna().sum())

working_data['RECP3'] = working_data['RECP3'].apply(lambda x: 1 if x=='X' else 0)
print(working_data[['RECP3']].value_counts())
print('NaNs:', working_data['RECP3'].isna().sum())

working_data['RECPGVG'] = working_data['RECPGVG'].apply(lambda x: 1 if x=='X' else 0)
print(working_data[['RECPGVG']].value_counts())
print('NaNs:', working_data['RECPGVG'].isna().sum())

working_data['RECSWEEP'] = working_data['RECSWEEP'].apply(lambda x: 1 if x=='X' else 0)
print(working_data[['RECSWEEP']].value_counts())
print('NaNs:', working_data['RECSWEEP'].isna().sum())
```

#### MDMAUD
Major Donor Matrix code, RFA feature

```python
# Major donor flag (binary)
working_data['MDMAUD_BIN'] = working_data['MDMAUD'].apply(lambda x : 0 if x == 'XXXX' else 1)
display(working_data['MDMAUD_BIN'].value_counts())
print('NaNs:', working_data['MDMAUD_BIN'].isna().sum())
```

```python
# RECENCY. We keep it categorical
working_data['MD_recency_giving'] = working_data['MDMAUD'].apply(lambda x : x[0])
working_data['MD_recency_giving'].replace({'C':'current',
                                        'L':'lapsed',
                                        'I':'inactive',
                                        'D':'dormant',
                                        'X':None}, inplace = True)
display(working_data['MD_recency_giving'].value_counts())
print('NaNs:', working_data['MD_recency_giving'].isna().sum())
```

```python
# FREQUENCY.
working_data['MD_frequency_giving'] = working_data['MDMAUD'].apply(lambda x: x[1])
working_data['MD_frequency_giving'][working_data['MD_frequency_giving']!='X'].hist();
```

```python
# We represent ordinal frequency as numeric value,
# 0 for minimum frequency (1 gift), 0.5 for 2-4, 1 for 5+ gifts 
working_data['MD_frequency_giving'].replace({'1':0,
                                           '2':0.5,
                                           '5':1,
                                           'X':None}, inplace = True)
display(working_data['MD_frequency_giving'].value_counts())
print('NaNs:', working_data['MD_frequency_giving'].isna().sum())
```

```python
# AMOUNT. We represent ordinal amount as numeric value,
# 0 for low amount (< 100$), 1 for top amounts (>1000$)
working_data['MD_amount_giving'] = working_data['MDMAUD'].apply(lambda x : x[2])
working_data['MD_amount_giving'].replace({'L':0,'C':1/3,'M':2/3,'T':1,'X':None}, inplace = True)

print('NaNs:', working_data['MD_amount_giving'].isna().sum())
print(working_data['MD_amount_giving'].value_counts())
working_data['MD_amount_giving'].hist();
```

#### DOMAIN
1st byte for urban level, 2nd for Socio-Economic status of neighbourhood

```python
print('NaNs:', sum(working_data['DOMAIN'].isna()))
working_data.DOMAIN.value_counts()
```

```python
# AMOUNT. We represent ordinal urbanization level as numeric value,
# 0 for rural area, 1 for urban area

working_data['urban_level'] = working_data.DOMAIN.apply(lambda x: x[0]).replace({'U':1,
                                                                         'C':0.75,
                                                                         'S':0.5,
                                                                         'T':0.25,
                                                                         'R':0,
                                                                         ' ': None})
print('NaNs:', sum(working_data['urban_level'].isna()))
print(working_data['urban_level'].value_counts())
working_data['urban_level'].hist();
```

```python
def SES_mapping(x):
    map_dict_urb = {'1' : 1,
                       '2' : 2/3,
                       '3' : 1/3,
                       '4' : 0}
    map_dict_nurb = {'1' : 1,
                    '2' : 0.5,
                    '3' : 0}
    if x == ' ':
        return None
    if x[0] == 'U':
        return map_dict_urb[x[1]]
    else:
        return map_dict_nurb[x[1]]
    
working_data['neigh_soc_econ_status'] = working_data.DOMAIN.apply(lambda x: SES_mapping(x))
print('NaNs:', sum(working_data['neigh_soc_econ_status'].isna()))
working_data['neigh_soc_econ_status'].value_counts()
```
